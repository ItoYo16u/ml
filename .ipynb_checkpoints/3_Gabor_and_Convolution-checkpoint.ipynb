# 課題03: ガボールフィルタ・畳み込み

(1) ガボールフィルタを作成し、表示してください。  
ガボールフィルタを作成するためには、`skimage.filters.gabor_kernel`を使用してください。この関数は複素数の値を返しますが、実数部分だけを使用してください。  
なお、関数の引数には次の値を使用してください。  
`
frequency = 0.4  
theta = 0  
sigma_x = 1.5
sigma_y = 1.5    
`  
引数の説明は公式ドキュメントに記載されています。  
http://scikit-image.org/docs/dev/api/skimage.filters.html#skimage.filters.gabor_kernel  

(2) 周波数 (frequency) を 0.2, 0.4, 1.5 に設定し、他のパラメータは (1) と同じものを使用してガボールフィルタを作成し、それぞれ表示してください。

(3) 方向 (orientation) を 0, π/4, 2π/4, 3π/4 に設定し、他のパラメータは (1) と同じものを使用してガボールフィルタを作成し、それぞれ表示してください。

(4) 配布した zebura.png を白黒画像として読み込んでください（二次元の配列として読み込む必要があります。下で紹介する関数を参照してください）。

(5) 方向が 0, π/4, 2π/4, 3π/4 のガボールフィルタをシマウマの画像に適用して、それぞれ結果を表示してください。それぞれの方向のエッジが強調されているはずです。

(6) 方向が 0, π/4, 2π/4, 3π/4 のガボールフィルタをシマウマの画像に適用したあと、それぞれの結果を足し合わせ、表示してください。縦、横、斜めのエッジが強調された画像を足し合わせたため、ほぼ全てのエッジが強調されているはずです。  
注意:   
`skimage.filters.gabor_kernel`を使用するとき、角度 (angle) の値によってガボールフィルタのサイズが変わることがあり、その場合結果の画像のサイズが異なり、足し合わせられません。  `scipy.signal.convolve2d`を引数 mode = same として使用すれば、結果の画像のサイズは常に入力画像と同じになり、足し合わせることが可能になります。


(7) 画像のエッジを検出する方法は多くありますが、その一つに Scharr フィルタがあります。Scharr フィルタは画像の微分係数の大きさを計算します。一般に、画素値が大きく変化するとき（はっきりとしたエッジが存在するとき）、微分係数の絶対値も大きくなります。Scharr フィルタは次で定義されます:   
np.array([[ -3-3j, 0-10j,  +3 -3j], [-10+0j, 0+ 0j, +10 +0j], [ -3+3j, 0+10j,  +3 +3j]])  
シマウマの画像に Scharr フィルタを適用して、結果を表示してください。Scharr フィルタを適用した後、絶対値をとることを忘れないでください。

# Homework 03: Gabor filter, Convolution

(1) Generate a gabor filter and display it.  
Please use "skimage.filters.gabor_kernel" to obtain it.   
This function returns values in the form of complex numbers, so please use only the real part.  
Please use the following paramters for "skimage.filters.gabor_kernel" :   
`
frequency = 0.4  
theta = 0  
sigma_x = 1.5
sigma_y = 1.5  
`

The arguments are explained in this site.  
http://scikit-image.org/docs/dev/api/skimage.filters.html#skimage.filters.gabor_kernel  

 
(2) Generate gabor filters with frequencies, 0.2, 0.4, 1.5 (Other parameters should be the same as Q1).

(3) Generate gabor filters with orientation 0, π/4, 2π/4, 3π/4 (Other parameters should be the same as Q1).


(4) Please load the zebra.png image as a gray scale image (you need to convert it to 2 dimensional vector, please see functions below)
 
(5) Display the result that you apply a gabor filter with orientation 0, π/4, π/2, 3π/4  to the image of zebra.   
(Edges in each direction should be emphasized respectively.)
 
(6) Apply Gabor filters with orientation 0, π/4, 2π/4, 3π/4 respectively, and display the combined images of the results.  
Since we sum the results of emphasis of vertical, horizontal, and diagonal edges, almost all edges should be emphasized.  
Note:  
When using "skimage.filters.gabor_kernel", the size of the Gabor filter may differ depending on the angle, and the size of the result of the convolution may differ and we may not be able to sum them. If you use "scipy.signal.convolve2d" with the argument "mode = 'same'", the size of the input image and the size of the output image will be the same, even if the filter size is different. This way, you can get the same size output

(7) There are many ways to detect edges in an image, one way is to apply Scharr operator to the image. The Scharr operator generates the gradient magnitude of the image. In general, there are sharp changes in intensity on edges, this corresponds to high magnitude of gradient in intensity. The Scharr operator is defined as:
np.array([[ -3-3j, 0-10j,  +3 -3j], [-10+0j, 0+ 0j, +10 +0j], [ -3+3j, 0+10j,  +3 +3j]]).
Display the result that you apply Scharr operator to the image of zebra. (Remember to take the absolute value of the filtered image.)
